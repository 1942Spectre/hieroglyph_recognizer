```python
import tensorflow as tf
from tensorflow.keras.models import Model,Sequential,load_model
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pydot
```

```python
def get_images(directory,input_size):
    
    images = []
    
    for image in os.listdir(directory):
        
        
        images.append(np.asarray(Image.fromarray(np.array(plt.imread(directory + "/" + image))).resize(input_size))/255)
        
    return np.array(images)
```

```python
images = get_images("EgyptImages/",(224,224))
```

```python
limages = images[0:3]
```

```python
rimages = images[3:]
```

```python
vgg_19 = load_model("vgg-19.h5")
```

```python
encoder = Sequential()
encoder.add(vgg_19.layers[0])
encoder.add(vgg_19.layers[1])
encoder.add(vgg_19.layers[2])
encoder.add(vgg_19.layers[3])
encoder.add(vgg_19.layers[4])
encoder.add(vgg_19.layers[5])
encoder.add(vgg_19.layers[6])
```

```python
for i in range(0,128):

    plt.imshow(encoder.predict(limages)[1][:,:,i],cmap="Greys")
    plt.show()
```
